# Count plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var =sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data=var,hue="smoker",palette='Dark2',saturation=0.1)
plt.show()
```

# Violin plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,palette='Dark2')
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],inner="quart") #"box","quartile",'point','stick'
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True)
plt.show()
```

# pair plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue='sex')
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],x_vars=["total_bill","tip"])  #y_vars
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg") #scatter,kde,hist,reg
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],kind="reg",diag_kind='hist') 
plt.show()
```

# Box plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var=sns.load_dataset('tips')
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',data=var,hue="sex",palette='dark:g')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day',y='total_bill',data=var,hue="sex",order=["Fri","Sun","Sat","Thau"],
            showmeans=True,linewidth=2, palette='plasma')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='total_bill',y='day',data=var,hue="sex",order=["Fri","Sun","Sat","Thau"],
            showmeans=True,linewidth=2, palette='plasma')
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x=var["total_bill"])
plt.show()
```
