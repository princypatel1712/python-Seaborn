# Histogram plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data=sns.load_dataset("penguins")
data
```

```python
sns.displot(data["bill_length_mm"])
plt.show()
```

```python
sns.displot(data["bill_length_mm"],bins=[35,40,45,50,55,60])
plt.show()
```

```python
sns.displot(data["bill_length_mm"],bins=[35,40,45,50,55,60],kde=True,rug=True,color="g")
plt.show() 
```

```python
sns.displot(data["bill_length_mm"],kde=True,rug=True,color="g",log_scale=True)
plt.show() 
```

# Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=sns.load_dataset("penguins").head(40)
data
```

```python
sns.barplot(x=data.island,y=data.bill_length_mm)
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex")
plt.show()
```

```python
order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",order=order_1)
plt.show()
```

```python
order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",order=order_1,
           hue_order=["Female","Male"],ci=100)
plt.show()
```

```python
order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",order=order_1,
           hue_order=["Female","Male"],ci=100,n_boot=2)
plt.show()
```

```python
order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=data,hue="sex",orient="h")
plt.show()
```

```python
order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",order=order_1,
           hue_order=["Female","Male"],orient="v",palette='ice')
plt.show()
```

```python
order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",order=order_1,
           hue_order=["Female","Male"],orient="v",palette='icefire',saturation=0.2)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=data, hue="sex", order=order_1, err_kws={'color': 'b'})
```

```python
order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",order=order_1,
           hue_order=["Female","Male"],orient="v",palette='icefire',saturation=0.2,
            errcolor="b",capsize=0.5)
plt.show()
```

```python
sns.set(style="darkgrid")

order_1 =["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",order=order_1,
           hue_order=["Female","Male"],orient="v",palette='icefire',saturation=0.2,
            errcolor="b",alpha=0.9)
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var,hue="sex",style="sex",sizes=(20,40))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var,hue="sex",style="sex",sizes=(20,40),
               palette="icefire",alpha=0.5)
plt.show()
```

```python
m={"Male":"*","Female":"o"}

sns.scatterplot(x="bill_length_mm",
                y="bill_depth_mm",data=var,hue="sex",style="sex",sizes=(20,40),
               palette="icefire",markers=m)
plt.show()
```
